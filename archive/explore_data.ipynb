```python
import numpy as np
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import pyautogui
```

```python
data_path = Path("data/processed")
csv_files = list(data_path.glob("*.csv"))

data = [pd.read_csv(f) for f in csv_files]
# data = [df.interpolate(method="linear", limit_direction="both").iloc[:1800] for df in data]
data = [df.iloc[:1800] for df in data]
```

```python
# get screen size
screen_width, screen_height = pyautogui.size()
print(screen_width, screen_height)
```

```python
f, ax = plt.subplots()
ax.set(xlim=(0, screen_width), ylim=(0, screen_height), aspect="equal")

for r in range(len(data)):
    ax.scatter(data[r]["cursor_x"] *1.5, data[r]["cursor_y"] * 1.5, c=data[r]["cursor_y"], cmap="bwr")
```

```python
for i in np.arange(1800):
    x, y = data[0].iloc[i][["cursor_x", "cursor_y"]]
    pyautogui.moveTo(x, y, duration=0.12, _pause=False)
```
